```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from dbr import *
import cv2
import numpy as np
from IPython.display import display, Image
import ipywidgets as widgets
import threading
import PIL
from PIL import ImageEnhance, ImageFilter
import PIL.ImageOps
from pyzxing import BarCodeReader
from ipywidgets import interact
import ipywidgets as widgets
from scripts.helper_functions import preprocess, text_decompose
```

### СТРИМ С ВЕБ КАМЕРЫ И ЗАХВАТ ИЗОБРАЖЕНИЯ

```python
stopButton = widgets.ToggleButton(
    value=False,
    description='Stop',
    disabled=False,
    button_style='danger',
    tooltip='Description',
    icon='square'
)


def view(button):
    cap = cv2.VideoCapture(0)
    display_handle=display(None, display_id=True)
    i = 0
    while True:
        _, frame = cap.read()
        _, image = cv2.imencode('.jpeg', frame)
        display_handle.update(Image(data=image.tobytes()))
        if stopButton.value==True:
            view.frame = frame
            cap.release()
            display_handle.update(None)

display(stopButton)
thread = threading.Thread(target=view, args=(stopButton,))
thread.start()
```

### ПРЕДОБРАБОТКА ИЗОБРАЖЕНИЯ И СОХРАНЕНИЕ В ФАЙЛ

```python
image = preprocess(view.frame, contrast_value=2)
image.save('temp/box_cropped.png')

fig, ax = plt.subplots(1,2, figsize=(10, 10))
ax[0].imshow(view.frame, cmap='gray')
ax[0].title.set_text('До обработки')
ax[1].imshow(image, cmap='gray')
ax[1].title.set_text('После обработки')
```
